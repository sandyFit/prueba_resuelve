# Accediendo al dataset

# Checando datos faltantes

```python
missing_data = df.isnull()
missing_data.head(5)
```

# Listar datos faltantes

```python
for column in missing_data.columns.values.tolist():
        print(column)
        print(missing_data[column].value_counts())
        print("")
```

# Reemplazar datos faltantes con NaN

```python
df.replace(" ", np.nan, inplace = True)
```

```python
df.head()
```

# Imputar valores faltantes
En la columna de historial crediticio hay 47 valores faltantes, un número considerable que si se eliminan podrían sesgar el modelo final, por eso se elije aquí la imputación con el más frecuente

```python
df['Historial_crediticio'].fillna(df['Historial_crediticio'].mode()[0], inplace=True)
df['Historial_crediticio'] = df['Historial_crediticio'].astype(int)
```

```python
df.head()
```

# Cambiar tipos de datos a datos numéricos y corregir Area_urbana

```python
df['Estatus_prestamo'] = df['Estatus_prestamo'].map({'Si': 1, 'No': 0})
df['Educacion'] = df['Educacion'].map({'Graduado': 1, 'No Graduado': 0})
df['Area_vivienda'] = df['Area_vivienda'].replace({'Urabana': 'Urbana'})
```

```python
df.sample(3)
```

```python
df['Dependientes'].fillna(df['Dependientes'].mode()[0], inplace=True)
df['Dependientes'].replace('3+', 4, inplace=True)
df['Dependientes'] = df['Dependientes'].astype(int)
df.sample(3)
```

```python
df['Casado'].fillna(df['Casado'].mode()[0], inplace=True)
df['Casado'] = df['Casado'].map({'Si': 1, 'No': 0})
```

```python
df['Salario'] = df['Salario'].astype('float')
```

```python
df.sample(5)
```

# Guardando dataset corregido

```python
df.to_csv('data.csv')
```
